### Data Preprocessing
1. Initialize
               a. pandas - pd functions
               b. numpy - numpy functions, np.reshape
               c. csv - import and export csv files

```python
import pandas as pd
import numpy as np
import csv
```

 2. Dropped
                a. Unnamed: 0 - index of .csv file
                b. Company - unnecessary
                c. Product - unnecessary (name)
                d. Weight - unnecessary
                e. Inches

```python
data = pd.read_csv("laptops.csv", header=0)
data.drop(['Unnamed: 0', 'Company', 'Product', 'Weight', 'Inches'], axis = 1, inplace = True)
```

```python
data
```

### TypeName

Convert entries into numerical data.

```python
list(data['TypeName'].unique())
```

```python
temp = list(data['TypeName'].unique())
data.TypeName = data.TypeName.replace(temp, [1,2,3,4,5,6])
```

```python
data
```

### OpSys

preprocess entries on OpSys column.

```python
list(data['OpSys'].unique())
```

Convert the entries on OpSys into numerical data and reduce to only Windows(1), Mac(2), Linux(3), and Others(4)

```python
data.OpSys = data.OpSys.replace({'Windows 10':1, 'Windows 10 S':1, 'Windows 7':1, 'macOS':2, 'Mac OS X':2, 'Linux':3, 'Android':4, 'Chrome OS':4, 'No OS': 4})
list(data['OpSys'].unique())
```

### RAM

Strip the unit of memory. All entries are GB.

```python
data['Ram'] = data['Ram'].str.rstrip('GB').astype(float)
```

### Price

Change Price entries to range. (799 and below Low; 800-1499 Average; 1500 and above High)

Change the Label of the column from Price_euros to Price

```python
copy = data.Price_euros.copy()
data.Price_euros = data.Price_euros.mask(copy<800, 'Low')
data.Price_euros = data.Price_euros.mask((copy<1500)&(copy>799), 'Average')
data.Price_euros = data.Price_euros.mask(copy>1499, 'High')
data.rename(columns={'Price_euros':'Price'},inplace = True)
```

### CPU
Convert entries to numerical.

```python
list(data['Cpu'].unique())
```

```python
data = data[~data['Cpu'].str.contains('Samsung')]

data.loc[data['Cpu'].str.contains('i3', na = False), 'Cpu'] = 1
data.loc[data['Cpu'].str.contains('i5', na = False), 'Cpu'] = 2
data.loc[data['Cpu'].str.contains('i7', na = False), 'Cpu'] = 3
data.loc[data['Cpu'].str.contains('Atom', na = False), 'Cpu'] = 4
data.loc[data['Cpu'].str.contains('Celeron Dual Core', na = False), 'Cpu'] = 5
data.loc[data['Cpu'].str.contains('Celeron Quad Core', na = False), 'Cpu'] = 6
data.loc[data['Cpu'].str.contains('Pentium Dual Core', na = False), 'Cpu'] = 7
data.loc[data['Cpu'].str.contains('Pentium Quad Core', na = False), 'Cpu'] = 8
data.loc[data['Cpu'].str.contains('Xeon', na = False), 'Cpu'] = 9
data.loc[data['Cpu'].str.contains('Core M', na = False), 'Cpu'] = 10
data.loc[data['Cpu'].str.contains('AMD', na = False), 'Cpu'] = 11
list(data['Cpu'].unique())
```

## Classification

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
```

```python
clean = pd.read_csv("cleaned.csv")
xHeader = list(data)
xHeader.remove('Price')
yHeader = ['Price']

list(data)
```

```python
x = pd.DataFrame(columns=xHeader)
x.TypeName = clean.TypeName
#x.Inches = clean.Inches
x.ScreenResolution = clean.ScreenResolution
x.Cpu = clean.Cpu
x.Ram = clean.Ram
x.Memory = clean.Memory
x.Gpu = clean.Gpu
x.OpSys = clean.OpSys

y = pd.DataFrame(columns=yHeader)
y.Price = clean.Price
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=400)
gnb = GaussianNB()
gnb = gnb.fit(x_train, np.ravel(y_train, order = 'C'))
y_pred = gnb.predict(x_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

np.ravel is used to flatten the array.

```python
print(metrics.classification_report(y_test, y_pred))
print(metrics.confusion_matrix(y_test,y_pred))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=400)
mnb = MultinomialNB()
mnb = mnb.fit(x_train, np.ravel(y_train, order = 'C'))
y_pred = mnb.predict(x_test)

accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

```python
print(metrics.classification_report(y_test, y_pred))
print(metrics.confusion_matrix(y_test,y_pred))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=400)
bnb = BernoulliNB()
bnb = bnb.fit(x_train, np.ravel(y_train, order = 'C'))
y_pred = bnb.predict(x_test)

accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

```python
print(metrics.classification_report(y_test, y_pred))
print(metrics.confusion_matrix(y_test,y_pred))
```
